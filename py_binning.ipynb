# Manual Bins

```python
import numpy as np
import matplotlib.pyplot as plt
from mycolorpy import colorlist as mcp
```

```python
rng = np.random.default_rng()
```

```python
center = 5
std = 1
num_std = 5
num_instances = 500
grain = 0.25

x = rng.normal(center, std, num_instances)

x_min = center - num_std * std
x_max = center + num_std * std

bins = np.linspace(x_min, x_max, int((x_max - x_min) / grain) +1 )

counts = np.zeros_like(bins)
i = np.searchsorted(bins, x)
np.add.at(counts, i, 1)

plt.style.use('dark_background')
plt.plot(bins, counts, drawstyle='steps')
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots()
ax.scatter(bins, counts, c='#ffaa55', s=10);   
plt.title('#ffaa55')
plt.xticks([]); plt.yticks([]); 
plt.xlabel(""); plt.ylabel("");       
ax.spines[:].set_visible(False)
```
